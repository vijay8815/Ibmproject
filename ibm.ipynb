```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"/home/subramaniyam/Downloads/rainfall.csv")
```

```python
df
```

```python
df2=df[df['YEAR']==2015]
```

```python
df2
```

```python
df3=df2[['SUBDIVISION','ANNUAL']]
```

```python
df3
```

```python
width1 = 15.
height1 = 5.
width_height_1 = (width1, height1)
plt. figure(figsize=width_height_1)
chart=plt.bar(df3['SUBDIVISION'],df3['ANNUAL'])
plt.xticks(rotation=90)
plt.title("RAINFALL IN INDIA IN THE YEAR OF 2015",fontsize=30)
plt.xlabel("STATES",fontsize=20)
plt.ylabel("RAINFALL",fontsize=20)
chart[31].set_color('red')
chart[0].set_color('red')
chart[1].set_color('red')
plt.legend(['TOP 3 STATES'])
```
